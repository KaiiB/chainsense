```python
# 1) Load checkpoint-2 artifacts
analysis_dir = REPO_ROOT / "data" / "analysis"

metrics_txt = (analysis_dir / "rep_a_metrics.txt").read_text(encoding="utf-8")
sizes = pd.read_csv(analysis_dir / "rep_a_cluster_sizes.csv")
labels = pd.read_csv(analysis_dir / "rep_a_cluster_labels.csv")
profile = pd.read_csv(analysis_dir / "rep_a_cluster_profile_median.csv")
clusters = pd.read_csv(analysis_dir / "rep_a_clusters.csv")
pca = pd.read_csv(analysis_dir / "rep_a_pca.csv")

print(metrics_txt)
display(sizes)
display(labels)
```

```python
# 2) Cluster size bar chart
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.bar(sizes["cluster"].astype(str), sizes["count"])
plt.title("Rep A Cluster Sizes")
plt.xlabel("Cluster")
plt.ylabel("Wallet Count")
plt.grid(axis="y", alpha=0.25)
plt.show()
```

```python
# 3) PCA scatter by cluster
plot_df = pca.merge(clusters, on=["wallet", "window_start", "window_end"], how="inner")

plt.figure(figsize=(7,5))
for c in sorted(plot_df["cluster"].unique()):
    s = plot_df[plot_df["cluster"] == c]
    plt.scatter(s["pc1"], s["pc2"], s=18, alpha=0.7, label=f"Cluster {c}")
plt.title("PCA Projection (PC1 vs PC2)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend()
plt.grid(alpha=0.2)
plt.show()
```

```python
# 4) Cluster profile table with labels
profile_labeled = profile.merge(labels, on="cluster", how="left")

key_cols = [
    "cluster", "provisional_label",
    "tx_count", "tx_out_count", "tx_in_count",
    "unique_counterparties_out", "unique_counterparties_in",
    "eth_sent_total", "eth_recv_total", "eth_net_flow",
    "total_gas_used", "pct_zero_value_txs"
]
key_cols = [c for c in key_cols if c in profile_labeled.columns]
display(profile_labeled[key_cols].round(4))
```

```python
# 5) Feature heatmap-style view (normalized cluster medians)
feat_cols = [c for c in profile.columns if c != "cluster"]
M = profile.set_index("cluster")[feat_cols].copy()
M_z = (M - M.mean()) / M.std(ddof=0).replace(0, np.nan)
M_z = M_z.fillna(0)

plt.figure(figsize=(12,4))
im = plt.imshow(M_z.values, aspect="auto")
plt.colorbar(im, label="z-score (cluster median)")
plt.title("Cluster Median Feature Heatmap (Normalized)")
plt.yticks(range(len(M_z.index)), [f"Cluster {c}" for c in M_z.index])
plt.xticks(range(len(M_z.columns)), M_z.columns, rotation=90)
plt.tight_layout()
plt.show()
```

```python
# 6) Top differentiating features per cluster (printout)
global_med = profile[feat_cols].median()
global_std = profile[feat_cols].std().replace(0, np.nan)
z = (profile.set_index("cluster")[feat_cols] - global_med) / global_std
z = z.fillna(0)

for c in z.index:
    top = z.loc[c].abs().sort_values(ascending=False).head(5)
    print(f"\nCluster {c} top differentiating features:")
    for feat in top.index:
        print(f"  - {feat}: z={z.loc[c, feat]:.2f}")
```
